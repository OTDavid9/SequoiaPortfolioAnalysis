# Exploratory Data Analysis for Scrapped Data of Sequoia Capital

## Import Libraries

```python
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
%matplotlib inline
```

## Load the Dataset

```python
# Load the dataset
df = pd.read_csv("sequoia_companies.csv")

# Display the first few rows and check the structure
df.head()
```

```python
df.info()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Remove extra spaces from column names
df.columns = df.columns.str.strip()
```

```python
df = df.fillna('Unknown')
df.isnull().sum()
```

```python
df.shape
```

## FEATURE ENGINEERING

```python
# Engineer a feature to count the number of partners
df['Number of Partners'] = df['Partners'].apply(lambda x: len(x.split(',')) if x != 'Unknown' else 0)
```

```python
# Engineer a feature to rate the current stage from lowest to highest
stage_ranking = {
    'Unknown': 1,
    'Pre-Seed/Seed': 2,
    'Early': 3,
    'Growth': 4,
    'IPO': 5,
    'Acquired': 6
}
df['Current Stage Rating'] = df['Current Stage'].apply(lambda x: stage_ranking[x])
```

```python
# Engineer a feature to extract first year of parntership

# Function to extract the year
def extract_year(stage):
    match = re.search(r'\((\d{4})\)', stage)
    return int(match.group(1)) if match else None

# Apply the function to extract the year
df['First Year Partnered'] = df['First Partnered'].apply(extract_year)

# Convert to datetime with only the year component
df['First Year Partnered'] = pd.to_datetime(df['First Year Partnered'], format='%Y')

# Display the DataFrame showing only the year component
df['First Year Partnered'] = df['First Year Partnered'].dt.year
```

```python
# Function to extract the year
def extract_year(stage):
    match = re.search(r'\((\d{4})\)', stage)
    return 2024 - (int(match.group(1))) if match else None

df['Age Since First Partnered'] = round(df['First Partnered'].apply(extract_year))
```

```python
# Function to extract the partner name
def extract_partner_name(stage):
    match = re.match(r'(.+?) \(\d{4}\)', stage)
    if match:
        return match.group(1)
    else:
        return "Unknown"  # Return None for unexpected stages

# Apply the function
df['First Partnered Stage'] = df['First Partnered'].apply(extract_partner_name)
```

```python
# Define custom stage ranking dictionary
stage_ranking

df['First Partnered Stage Rating'] = df['First Partnered Stage'].apply(lambda x: stage_ranking[x])
```

```python
df["Stage Difference"] = df['Current Stage Rating'] - df['First Partnered Stage Rating']

df["Performance Comparison"] = (df['Current Stage Rating'] - df['First Partnered Stage Rating']) / df['First Partnered Stage Rating']  * 100
df["Performance Comparison (%)"] =df["Performance Comparison"].round(2)
```

```python
df.head()
```

```python
# Calculate counts of each category
stage_counts = df['First Partnered Stage'].value_counts()

# Order the categories by count
order = stage_counts.index

# Distribution of Current Stage (horizontal bar plot)
plt.figure(figsize=(12, 6))
ax = sns.countplot(y='First Partnered Stage', data=df, orient='h', order=order)
plt.grid()

plt.title('Distribution of First Partnered Stage of Companies')

# Annotate each bar with the count
for p in ax.patches:
    ax.annotate(f'{p.get_width()}', 
                (p.get_width(), p.get_y() + p.get_height() / 2), 
                ha='left', va='center', xytext=(5, 0), textcoords='offset points')

plt.xlabel('Count')
plt.ylabel('First Partnered Stage')
plt.savefig("First Partnered Stage.png")
plt.show()
```

```python
# Calculate counts of each category
stage_counts = df['Current Stage'].value_counts()

# Order the categories by count
order = stage_counts.index

# Distribution of Current Stage (horizontal bar plot)
plt.figure(figsize=(12, 6))
ax = sns.countplot(y='Current Stage', data=df, orient='h', order=order)
plt.grid()

plt.title('Distribution of Current Stage of Companies')

# Annotate each bar with the count
for p in ax.patches:
    ax.annotate(f'{p.get_width()}', 
                (p.get_width(), p.get_y() + p.get_height() / 2), 
                ha='left', va='center', xytext=(5, 0), textcoords='offset points')

plt.xlabel('Count')
plt.ylabel('Current Stage')
plt.savefig("Current Stage.png")
plt.show()
```

```python
# Calculate counts of each category
stage_counts = df['Number of Partners'].value_counts()

# Order the categories by count
order = stage_counts.index

# Distribution of Current Stage (horizontal bar plot)
plt.figure(figsize=(12, 6))
ax = sns.countplot(y='Number of Partners', data=df, orient='h', order=order)
plt.grid()

plt.title('Distribution of Partners')

# Annotate each bar with the count
for p in ax.patches:
    ax.annotate(f'{p.get_width()}', 
                (p.get_width(), p.get_y() + p.get_height() / 2), 
                ha='left', va='center', xytext=(5, 0), textcoords='offset points')

plt.xlabel('Count')
plt.ylabel('Number of Partners')
plt.show()
```

```python
# Calculate percentages of each category
stage_counts = df['Stage Difference'].value_counts()
stage_percentages = stage_counts / len(df) * 100

# Create a new DataFrame for percentages
df_percentages = stage_percentages.reset_index()
df_percentages.columns = ['Stage Difference', 'Percentage']

# Order the categories by percentage
order = df_percentages['Stage Difference']

# Distribution of Stage Difference (horizontal bar plot)
plt.figure(figsize=(12, 6))
ax = sns.barplot(y='Stage Difference', x='Percentage', data=df_percentages, orient='h', order=order)

# Annotate each bar with the percentage
for p in ax.patches:
    ax.annotate(f'{p.get_width():.2f}%', 
                (p.get_width(), p.get_y() + p.get_height() / 2), 
                ha='left', va='center', xytext=(5, 0), textcoords='offset points')

plt.grid()
plt.title('Distribution of Stage Progress')
plt.xlabel('Percentage')
plt.ylabel('Stage Difference')
plt.savefig("Stage Progress.png")
plt.show()
```

```python
# Aggregate 'Number of Partners' and group by 'First Year Partnered'
grouped_df = df.groupby('First Year Partnered')['Number of Partners'].sum().reset_index()

# Plotting the aggregated data
plt.figure(figsize=(10, 6))
plt.plot(grouped_df['First Year Partnered'], grouped_df['Number of Partners'], marker='o', linestyle='-', color='blue')
plt.title('Total Number of Partners by First Year Partnered')
plt.xlabel('First Year Partnered')
plt.ylabel('Total Number of Partners')
plt.grid(True)
plt.show()
```

```python
# Aggregate performance of startups and grouped by partners
grouped_df = df.groupby('Number of Partners')['Performance Comparison (%)'].mean().reset_index()

# Plotting the aggregated data
plt.figure(figsize=(10, 6))
plt.plot( grouped_df['Number of Partners'],grouped_df['Performance Comparison (%)'], marker='o', linestyle='-', color='blue')
plt.title('Number of Partners vs Average Performance of Startups (%)')
plt.xlabel('Numbers of Partners')
plt.ylabel('Average Performance of Startups (%)')
plt.grid(True)
plt.savefig("Performance and Partners.png")
plt.show()
```

```python
# Aggregate performance of startups and grouped by partners
grouped_df = df.groupby('Age Since First Partnered')['Performance Comparison (%)'].mean().reset_index()

# Plotting the aggregated data
plt.figure(figsize=(10, 6))
plt.plot( grouped_df['Age Since First Partnered'],grouped_df['Performance Comparison (%)'], marker='o', linestyle='-', color='blue')
plt.title('Years of Existence vs Average Performance of Startups (%)')
plt.xlabel('Years of Existence')
plt.ylabel('Average Performance of Startups (%)')
plt.grid(True)
plt.savefig("Performance and Years of Existence.png")
plt.show()
```

```python
# Aggregate performance of startups and grouped by partners
grouped_df = df.groupby('Age Since First Partnered')['Stage Difference'].mean().reset_index()

# Plotting the aggregated data
plt.figure(figsize=(10, 6))
plt.plot( grouped_df['Age Since First Partnered'],grouped_df['Stage Difference'], marker='o', linestyle='-', color='blue')
plt.title('Years of Existence vs Stage Difference of Startups (%)')
plt.xlabel('Years of Existence')
plt.ylabel('Stage Difference')
plt.grid(True)
plt.savefig("Years and Stage Difference.png")
plt.show()
```
